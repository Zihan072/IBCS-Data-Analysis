```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
B2_6691 = pd.read_csv("./data/B2_100µM_6691.csv")
B2_6691
```

```python
# B2_6691.columns
# for col in B2_6691.columns:
#     print col
fig = plt.figure(figsize=(15,10),dpi=100)
ax=fig.add_subplot(1,1,1)
for col in B2_6691.columns:
    ax.plot(B2_6691.index,B2_6691[col],label=str(col))
#     ax.set_xlabel[str(col)]
plt.legend()
```

```python
fig = plt.figure(figsize=(15,10),dpi=100)
ax=fig.add_subplot(1,1,1)
for col in B2_6691.columns:
    ax.plot(B2_6691.index,B2_6691[col],label=str(col))
#     ax.set_xlabel[str(col)]
plt.legend()
ax.set_ylim(0,0.5)
ax.text(190, -0.05, '10s', fontsize=14)
ax.text(380, -0.05, '20s', fontsize=14)
plt.xlabel('Time(Frame)', fontsize=18)
plt.ylabel('Motion Index', fontsize=18)
plt.title('B2_100µM_6691', fontsize = 20)
plt.savefig("./picture/B2_100µM_6691", dpi=300)
```
